### Reto - Descenso de gradiente y regresión lineal

Implemente el código del descenso por gradiente para la regresión lineal.

$$ y_i = w_1 x_i + w_0 $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos con ruido
x = np.random.uniform(0, 1, 20)
noise_variance = 0.2
noise = np.random.randn(x.shape[0]) * noise_variance
y = x * 2 + noise

# Descenso de gradiente
learning_rate = 0.1
max_iterations = 1000
w0 = 0
w1 = 1

for i in range(max_iterations):
    y_predicted = w1 * x + w0
    gradient_w0 = -2 * np.mean(y - y_predicted)
    gradient_w1 = -2 * np.mean((y - y_predicted) * x)
    w0 -= learning_rate * gradient_w0
    w1 -= learning_rate * gradient_w1

# Imprimir los resultados
print("Parámetros finales:")
print("w0:", w0)
print("w1:", w1)

# Graficar los datos y la línea ajustada

plt.plot(x, y, 'o', label='y (con ruido)')
plt.plot(x, x*2, 'b-', label='f(x)')
plt.plot(x, w1*x + w0, 'r-', label='Ajuste lineal con descenso de gradiente')
plt.xlabel('$x$', fontsize=15)
plt.ylabel('$y$', fontsize=15)
plt.ylim([0, 2])
plt.title('inputs (x) vs targets (y)')
plt.grid()
plt.legend(loc=2)
plt.gcf().set_size_inches((10, 6))
plt.show()
```
